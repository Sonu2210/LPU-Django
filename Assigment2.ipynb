```python
"Q5."
n=int(input("Enter a number: "))
c=int(input("Press 1= Sum \n Press 2= Product"))
def calc_sum():
    s=0
    for i in range(1,n+1):
        s+=i
    print("Sum = ",s)
def calc_pro():
    p=1
    for i in range(1,n+1):
        p=p*(i)
    print("Product = ",p)
if(c==1):
    calc_sum()
elif(c==2):
    calc_pro()
else:
    print("Enter a valid choice !!!")
```

```python
"Q9."
def fact(number):
    factorial = 1
    while(number != 1):
        factorial = factorial*number
        number -=1
    return factorial
n = int(input("Enter a number:"))
print(f"Factorial of {n} is {fact(n)}")
```
